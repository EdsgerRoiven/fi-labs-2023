```python
import re
def format(filename):
  with open(filename, encoding = 'utf-8') as file:
    text_string = file.read()
    file.close()
  low = text_string.lower()
  update = re.sub(r'[^а-яё ]','',low)
  update = re.sub(' +', ' ', update)
  return update
```

```python
text = format('лаба2.txt')
print(text)
```

```python
alphabet = 'абвгдежзийклмнопрстуфхцчшщъыьэюя'

def letter_frequency(text, alphabet):
  letter_count = {}
  for elem in alphabet:
    letter_count[elem] = 0
  for letter in text:
    letter_count[letter] = letter_count[letter]+1
  letter_count = {}
  for letter in alphabet:
    letter_count[letter] = round((letter_count[letter])/len(text) , 5)
  return letter_count
```

```python
import random
def generate_key(length):
  key = ''
  i = 0
  for i in range(length):
    letter = random.choice(alphabet)
    key+=letter
  return key
```

```python
def encrypt(text, key):
  result = ''
  i = 0
  while i < len(text):
    encrypt_letter = text[i] + key[i%len(key)]
    result += encrypt_letter
    i = i + 1
  print(result)
```

```python
encrypt(text, generate_key(3))
```

```python
def create_blocks(text, block_length):
  new_texts = []
  i = 0
  while i < block_length:
    new_texts.append(text[i])
    i+=1
  for j in range(block_length, len(text)):
    new_texts[j%block_length]+=text[j]
    j+=1
  return new_texts
```

```python
result = create_blocks(text,3)
```

```python
def CIndex(alphabet, text):
  index = 0
  for letter in alphabet:
    summa = 0
    iter = 0
    while iter < len(text):
      if letter == text[iter]:
        summa+=1
      iter+=1
    index+=summa*(summa-1)
  index = index/(len(text)*(len(text)-1))
  return index
```

```python
print(CIndex(alphabet, text))
```
