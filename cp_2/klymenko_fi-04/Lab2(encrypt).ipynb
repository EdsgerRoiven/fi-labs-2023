```python
import re
def format(filename):
  with open(filename, encoding = 'utf-8') as file:
    text_string = file.read()
    file.close()
  low = text_string.lower()
  update = re.sub(r'[^а-яё ]','',low)
  update = re.sub(' +', ' ', update)
  return update
```

```python
text = format('лаба2.txt')
print(text)
```

```python
alphabet = 'абвгдежзийклмнопрстуфхцчшщъыьэюя'

def letter_frequency(text, alphabet):
  letter_count = {}
  for elem in alphabet:
    letter_count[elem] = 0
  for letter in text:
    letter_count[letter] = letter_count[letter]+1
  letter_count = {}
  for letter in alphabet:
    letter_count[letter] = round((letter_count[letter])/len(text) , 5)
  return letter_count
```

```python
import random
def generate_key(length):
  key = ''
  i = 0
  for i in range(length):
    letter = random.choice(alphabet)
    key+=letter
  return key
```

```python
key = generate_key(3)
print(key)
```

```python
def encrypt(text, key):
  result = ''
  i = 0
  while i < len(text):
    encrypt_letter = text[i] + key[i%len(key)]
    result += encrypt_letter
    i = i + 1
  print(result)
```

```python
encrypt(text, generate_key(3))
```
