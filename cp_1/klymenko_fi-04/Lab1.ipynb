```python
import re
def format(filename):
  with open(filename, encoding = 'utf-8') as file:
    text_string = file.read()
    file.close()
  low = text_string.lower()
  update = re.sub(r'[^а-яё ]','',low)
  update = re.sub(' +', ' ', update)
  return update
```

```python
alphabet_probel ='абвгдеёжзийклмнопрстуфхцчшщъыьэюя '
alphabet = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'

def frequency_monogram(text, alphabet):
  letter_count = {}
  for letter in alphabet:
    letter_count[letter] = 0
  for letter in text:
    if letter!=' ':
      letter_count[letter]+=1
  if ' ' in alphabet:
    letter_count[' '] = 83088
  count_sum = 0
  for key in letter_count:
    count_sum+=letter_count[key]
  for key in letter_count:
    letter_count[key] = letter_count[key]/count_sum
  return letter_count
```

```python
def frequency_bigram(text, alphabet):
  pair_letters_count = {}
  for first in alphabet:
    for second in alphabet:
      pair_letters_count[first+second] = 0
  iter = 0
  while iter < len(text)-1:
    if text[iter] + text[iter+1] in pair_letters_count.keys():
      key = text[iter] + text[iter+1]
      pair_letters_count[key] +=1
    iter = iter + 1
  summa = 0
  for key in pair_letters_count:
    summa+=pair_letters_count[key]
  for key in pair_letters_count:
    pair_letters_count[key] = round(pair_letters_count[key]/summa, 5)
  return pair_letters_count
```

```python
print(frequency_bigram(text, alphabet))
```
